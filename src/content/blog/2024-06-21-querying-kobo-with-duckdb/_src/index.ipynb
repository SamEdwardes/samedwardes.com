````python
import duckdb

%load_ext sql
conn = duckdb.connect('KoboReader.sqlite')
%sql conn --alias duckdb
%sql SET GLOBAL sqlite_all_varchar = true
````

````python
%config SqlMagic.displaylimit = None
````

````python
%%sql
SHOW TABLES;
````
